# Titanic analysys with Pandas

```python
import pandas as pd
```

```python
titanic = pd.read_csv("./archive/titanic_train.csv")
titanic.head(5)
```

```python
titanic.info()
```

```python
class_survival = titanic.groupby("Pclass")["Survived"].mean()
print(class_survival)
```

```python
price_mean_by_class = titanic.groupby("Pclass")["Fare"].mean()
price_mean_by_class
```

```python
fare_survival = titanic.groupby("Survived")["Fare"].mean()
fare_survival
```

```python
total_young_people_df = titanic[(titanic["Age"] < 18)]
total_young_people_df.groupby("Age")["Age"].count()
```

```python
# Now, let's filter out data for passengers who are under 18 and survived
titanic_young_survived_df = titanic[(titanic["Age"] < 18) & (titanic["Survived"] == 1)]
survived_children_count = titanic_young_survived_df.groupby("Age")["Survived"].count()
survived_children_count
```

##  The highest survival rate among young people under 18 in the Titanic 

```python
# Total young people count by age
total_young_people_count = total_young_people_df.groupby("Age")["Age"].count()

# Survived children count by age
survived_children_count = titanic_young_survived_df.groupby("Age")["Survived"].count()

# Calculate survival rate for each age
survival_rate_by_age = (survived_children_count / total_young_people_count) * 100
survival_rate_by_age
```

```python
# Find the age with the highest survival rate
age_with_highest_survival_rate = survival_rate_by_age.idxmax()
highest_survival_rate = survival_rate_by_age.max()

print(f"Age with highest survival rate: {age_with_highest_survival_rate}, Survival Rate: {highest_survival_rate}%")
```
