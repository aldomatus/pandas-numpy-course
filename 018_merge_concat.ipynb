```python
import pandas as pd 
import numpy as np
```

# Concat

```python
df_1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
df_1

```

```python
df_2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
})
df_2
```

```python
pd.concat([df_1, df_2], ignore_index=True)
```

```python
pd.concat([df_1, df_2], axis=1, ignore_index=True)
```

# Merge

```python
left_df = pd.DataFrame({
    "key": ["k0", "k1", "k2", "k3"], 
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
})

right_df = pd.DataFrame({
    "key": ["k0", "k1", "k2", "k3"], 
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
left_df.merge(right_df, on="key")
```

```python
left_df = pd.DataFrame({
    "key": ["k0", "k1", "k2", "k3"], 
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
})

right_df = pd.DataFrame({
    "key_2": ["k0", "k1", "k2", "k3"], # it change to key_2
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
left_df.merge(right_df, left_on="key", right_on="key_2")
```

```python
left_df = pd.DataFrame({
    "key": ["k0", "k1", "k2", "k3"], 
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
})

right_df = pd.DataFrame({
    "key_2": ["k0", "k1", "k2", np.nan], # changed k3 to nan
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
left_df.merge(right_df, left_on="key", right_on="key_2", how="inner") # how (left, right or inner)
```
